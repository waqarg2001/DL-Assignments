```python
import numpy as np
import pandas as pd
from tensorflow.keras import models,layers
data=pd.read_csv('/content/ionosphere_data.csv')
```

```python
data
```

```python
data.isna().sum()
```

```python
data.drop_duplicates(inplace=True)
```

```python
data
```

```python
data=data.sample(frac=1)
```

```python
data['label']=data['label'].map({'g':1,'b':0})
labels=data['label']
```

```python

data.drop('label',inplace=True,axis=1)
mean=data.mean(axis=0)
std=data.std(axis=0)
data-=mean
data/=mean
```

```python
data
```

```python
data.feature2.isna().sum()
```

```python
data['feature2']=0
```

```python
train_data=data[:210]
train_labels=labels[:210]

test_data=data[210:]
test_labels=labels[210:]
```

```python
model=models.Sequential()
model.add(layers.Dense(16,activation='relu',input_shape=(train_data.iloc[1].shape),kernel_initializer='normal'))
model.add(layers.Dropout(0.5))
model.add(layers.Dense(8,activation='relu',kernel_initializer='normal'))


model.add(layers.Dense(1,activation='sigmoid',kernel_initializer='normal'))
model.compile(optimizer='rmsprop',loss='binary_crossentropy',metrics=['accuracy'])
model.fit(train_data,train_labels,epochs=100,batch_size=1)
```

```python
model.evaluate(test_data,test_labels)
```
